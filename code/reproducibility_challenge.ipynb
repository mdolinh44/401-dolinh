```python
## Reproducibility Challenge
## Name: Max Dolinh

# pandas to load csv as data frame, numpy for distance formula
import pandas as pd
import numpy as np

starbucks = pd.read_csv('data/starbucks_locations.csv')


def return_nearest(n, longitude, latitude):
    # iterate through starbucks stores
    
    for store in starbucks:
        # calculate each starbucks store's distance from the input parameters
        starbucks['dlong'] = starbucks['Longitude'] - longitude
        starbucks['dlat'] = starbucks['Latitude'] - latitude
        starbucks['distance'] = np.sqrt(starbucks['dlong']**2 + starbucks['dlat']**2)

    # sort by closest stores to input parameters
    starbucks_sorted = starbucks.sort_values(by='distance', ascending = True)

    # print name of closest store(s) to input latitude and longitude
    return starbucks_sorted.head(n)

return_nearest(5, 40, 40)
```
